```python
import numpy as np
import random
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
class Ball_Frame:
    def __init__(self,):
        self.dataset_size = 5
        self.num_frames = 20
        self.frame_width = 100
        self.frame_height = 50
        self.step = 2

    def make_frame(self,label, init_x, init_y):
        frames = []
        next_y = init_y
        next_x = init_x

        for i in range(self.num_frames):

          frame = np.zeros((self.frame_height, self.frame_width))
          cv2.circle(frame, (next_y, next_x), 8, 255, -1) 

          next_x = random.randint(next_x - 2, next_x + 2)

          if label==1:
            next_y += self.step 

          else:
            next_y -= self.step

          frames.append(frame)

        return frames
    
    def generate_data(self,initial_state_x, random_num):

        if random_num < 0.5:
          label = 0
          initial_state_y = self.frame_width - 1

        else:
          label = 1
          initial_state_y = 1

        frames = self.make_frame(label, initial_state_x, initial_state_y)

        return frames, label
    def main(self,):
      
      for d in range(self.dataset_size):
          random_num = random.random()
          initial_state_x = random.randint(2, self.frame_height - 2)

          data, label = self.generate_data(initial_state_x, random_num)

          dataset.append(data)
          labels.append(label)
```

```python
labels = []
data = []
dataset = []
```

```python
# Convert to np.array
ball = Ball_Frame()
ball.main()
dataset = np.array(dataset)
labels = np.array(labels)
dataset.shape, labels.shape
```

```python
# Reshape dataset

dataset = dataset[..., np.newaxis]
labels = labels[..., np.newaxis]

dataset.shape, labels.shape
```
