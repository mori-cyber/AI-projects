```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import losses,models,layers,optimizers
from  tensorflow.keras.optimizers import  SGD,Adam
```

```python
train_dataset = pd.read_csv('/content/train.csv')
```

```python
test_dataset = pd.read_csv('/content/test.csv')
```

```python
train_dataset = train_dataset.replace(['female','male'],[0,1])
train_dataset = train_dataset.replace(['S','C','Q'],[0,1,2])
train_dataset = train_dataset.fillna(0)
```

```python
y_train = train_dataset[['Survived']]
x_train = train_dataset[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
Y_train = np.array(y_train)
X_train = np.array(x_train)
```

```python
model = models.Sequential([
                      layers.Dense(7,input_dim = 7, activation='sigmoid'),
                      layers.Dense(8, activation='relu'),
                      layers.Dense(4, activation='sigmoid'),
                      layers.Dense(2, activation='softmax')
])
```

```python
model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
loss=losses.sparse_categorical_crossentropy,
metrics=['accuracy'])
```

```python
output=model.fit(X_train,Y_train,epochs=200)
```

```python
class KNearestNeighbors:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X_train, Y_train):
        self.X_train = X_train
        self.Y_train = Y_train
        self.number_class = len(np.unique(Y_train))
        
    def nearestNeighbors(self, X_test):
        distance = np.sqrt(np.sum((X_test - self.X_train)**2, axis = 1))
        near_neighbor = np.argsort(distance)[0:self.k]
        return near_neighbor
    
    def predict(self, X_test):
        self.X_test = X_test
        y_predict=[]
        for i, test in enumerate(self.X_test):
            near_neighbor = self.nearestNeighbors(test)
            y_predict.append([np.argmax(np.bincount(self.Y_train[near_neighbor]))])
        return y_predict
    
    def evaluate(self, X_test, Y_test):
        self.X_test = X_test
        self.Y_test = Y_test
        y_predict=self.predict(self.X_test)
        evaluatation = (y_predict == self.Y_test).sum()/len(self.Y_test)
        return evaluatation, y_predict


```

```python
Y_train = Y_train.reshape(891,)
knn = KNearestNeighbors(k=5)
knn.fit(X_train,Y_train)
knn_prediction = knn.predict(x_test)
evaluatation, prediction = knn.evaluate(x_test, y_test)
print(' Accuracy of KNearestNeighbors is', evaluatation)
```

```python
class AdalineClassifier:
    def __init__(self):
        pass
    
    def fit(self, X_train, Y_train):
        self.w = np.matmul(inv(np.matmul(X_train.T, X_train)), np.matmul(X_train.T, Y_train))
        
    def predict(self, X_test):
        Y_predic = np.matmul(X_test, self.w)
        return Y_predic
    
    def evaluate(self, X_test, Y_test):
        self.X_test = X_test
        self.Y_test = Y_test
        Y_predic = np.matmul(self.X_test, self.w)
        predic = []
        for i, pred in enumerate(Y_predic):
            if pred > 0.5:
                predic.append([1])
            elif pred < 0.5:
                predic.append([0])
        evaluatation = ((predic == self.Y_test).sum())/len(self.Y_test)
        return evaluatation, predic
```

```python
from numpy.linalg import inv
model = AdalineClassifier()
model.fit(X_train, Y_train)
y_predic = model.predict(x_test)
evaluatation, prediction = model.evaluate(x_test, y_test)
print('Accuracy of Adaline is', evaluatation)
```

```python
class Perceptron:
    def __init__(self,learning_rate, n_iters):
        self.lr=learning_rate
        self.n_iters=n_iters
        self.w = np.random.rand(7,1)
        self.b = np.random.rand(1,1)
        self.Error = []
    def fit(self, X_train, Y_train):
        self.X_train = X_train
        self.Y_train = Y_train
        
        for i in range(self.X_train.shape[0]):
            y_pred = np.matmul(self.X_train[i], self.w) + self.b
            e = self.Y_train[i] - y_pred
            a = self.lr * self.X_train[i, :].T * e
            
            self.w += a.T
            self.b += self.lr * e

            Y_pred = np.matmul(self.X_train, self.w) + self.b
            error = np.mean(np.abs(self.Y_train - Y_pred))
            self.Error.append(error)
            np.save('weight', self.w)
            np.save('baias', self.b)
        return self.Error, self.w, self.b
            
    def predict(self,X_test):
        w = np.load('weight.npy')
        b = np.load('baias.npy')
        y_predict = np.matmul(X_test, w)+b
                    
        return y_predict
    

    
    def evaluate(self,X_test,Y_test):
        w = np.load('weight.npy')
        b = np.load('baias.npy')
        Y_pred = np.matmul(X_test, w) + b
        absolute_error = np.abs(Y_pred - Y_test)
        evaluation = np.mean(absolute_error)
        return evaluation
```
