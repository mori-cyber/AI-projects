```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_dataset = pd.read_csv('COVID-19-time-series-clean-complete.csv')
```

```python
covid_dataset.head(10)
```

```python

covid_dataset['Date'] = pd.to_datetime(covid_dataset['Date'])
covid_dataset['YearMonth'] = covid_dataset['Date'].map(lambda date : 100 * date.year + date.month)
```

```python
april_data = covid_dataset.query("YearMonth == 202004").reset_index(drop=True)
april_data.head(8)
```

```python
april_month_data = covid_dataset.query("YearMonth == 202004").reset_index(drop=True)
april_month_data.head(8)
```

```python
total_Confirmed = april_month_data.groupby('Country/Region')['Confirmed'].sum().reset_index()
```

```python
sorted_total_Confirmed = total_Confirmed.sort_values(by=['Confirmed'], ascending=False)
sorted_total_Confirmed.head(8)
```

```python
iran_country = covid_dataset.loc[(covid_dataset['Country/Region'] == 'Iran')]
iran_country.head()
```

```python
iran_covid19 = pd.DataFrame({'days' : iran_country['Date'], 'New deaths' : iran_country['New deaths'], 'New cases' : iran_country['New cases']})
```

```python
iran_covid19.plot.area(x='days')
# iran_covid19.plot(kind='area')
```
